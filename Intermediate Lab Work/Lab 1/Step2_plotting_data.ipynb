# Ploting data

In the cell below, keep numpy and load the matplotlib library

```python
import numpy as np
import matplotlib.pyplot as plot
```

Keep the cells that load the data 

```python

data_file = np.loadtxt("test_data_lab01.csv",delimiter=',',skiprows=2)
v = data_file[:,0]
dv = data_file[:,1]
i = data_file[:,2] 
di = data_file[:,3]
```

The cell below shows how to plot the data a graph
Apply a label to the vertical and horizonal axes
Apply an over all title  
Save the plot as a pdf ---> This will be useful for putting plots in your reports

```python
plot.errorbar(x=v,y=i,xerr=dv,yerr=di,fmt='o')
plot.xlabel('Applied Voltage (V)')
plot.ylabel('Measured Current (mA)')
plot.title("Measured Current vs. Applied Voltage")
plot.savefig("My-beautiful-data.pdf")
```
